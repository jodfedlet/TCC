```python
def pre_processing_tweets(tweet):
  tweet_list = str(tweet).lower().split()

  def remove_undesirable_words(tweet_list):
      return [remove_emphasis(word) for word in tweet_list if not word.startswith('@') and not word.startswith('#') and len(word) > 3]

  def remove_emphasis(word):
      return unidecode(word)

  def remove_stop_words(tweet_list):
      return ' '.join([word.lower() for word in tweet_list if word not in stop_words])

  def remove_url(tweet_without_sw):
      return re.sub(r"http\S+", "", tweet_without_sw)

  def remove_punctuation(tweet):
      return re.sub(r'[^\w\s]','',tweet)

  tweet_list = remove_undesirable_words(tweet_list)
  tweet_without_sw = remove_stop_words(tweet_list)
  tweet = remove_url(tweet_without_sw)
  return remove_punctuation(tweet)
```

```python
string_example = "@gen_heleno VAI BOTAR UM PIJAMA. E VAI PRA CASA DESCANSAR.  REZA E PEÇA A DEUS QUE PERDOE TODAS AS BARBARIDADES QUE VOCÊ COMETEU NO HAITI. EXISTEM ALMAS LÁ QUE NUNCA IRÃO PERDOAR AS SUAS ATITUDES. VOCÊ E DEUS SABEM DISTO."
```

```python
filtered_tweet = pre_processing_tweets(string_example.lower())
```
